```python
x=np.arange(0,10)
y=np.arange(11,21)
x
y
plt.plot(x,y,'b.--',linewidth=2,markersize=5)
plt.title("modern")
plt.xlabel("X axis")
plt.ylabel('Y axis')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("modern 1")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,'go--',linewidth=12,markersize=12)
```

```python
x=np.arange(0,5*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y,color='r')
plt.title("cosine wave")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
